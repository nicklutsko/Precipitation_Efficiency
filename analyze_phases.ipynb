```python
import numpy as np
import matplotlib.pylab as plt
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
%matplotlib inline
```

```python
#List of temperatures as floats and strings
temp1 = [  290., 292.5, 295., 297.5, 300., 302.5, 305., 307.5, 310.]
stemp1 = ["290", "292.5", "295", "297.5", "300", "302.5", "305", "307.5", "310"]

t = len(temp1)
```

```python
#Load data for water paths and integrate. Multiply by 1000 to make g/kg
qc = np.load( "data/small_domain_micro_qc.dat") * 1000. #liquid condensate
qi = np.load( "data/small_domain_micro_qi.dat") * 1000. #ice condensate

qr = np.load( "data/small_domain_micro_qr.dat") * 1000. #rain
qg = np.load( "data/small_domain_micro_qg.dat") * 1000. #graupel
qs = np.load( "data/small_domain_micro_qs.dat") * 1000. #snow

z = np.load( "data/small_domain_micro_z.dat")


qc = np.trapz( qc, z)
qi = np.trapz( qi, z)

qr = np.trapz( qr, z)
qg = np.trapz( qg, z)
qs = np.trapz( qs, z)
```

```python
def make_axis( a ):

    a.tick_params( axis = 'x', which = 'both', bottom="on", top = "off", labelbottom = "on", direction = 'out')                  
    a.tick_params( axis = 'y', which = 'both', bottom="off", top="off",  labelbottom="off", left="on", right="off", labelleft="on", direction = 'out')                                                                                        

    a.tick_params(length = 5, which = 'both')
    a.spines["top"].set_visible(False)    
    a.spines["right"].set_visible(False) 

    majorLocator = MultipleLocator(5)
    minorLocator = MultipleLocator(2.5)
    a.xaxis.set_major_locator(majorLocator)
    a.xaxis.set_minor_locator(minorLocator)

    plt.xlim([289., 311.])
    plt.xlabel( "SST [K]", fontsize = 14 )

    return 0
```

```python
#Plot!
fig = plt.figure( figsize = (15, 8) )
plt.subplots_adjust(left = 0.06, right = 0.97, bottom = 0.07, top = 0.97, hspace = 0.3, wspace = 0.3)


cs = [(31 / 255., 119 / 255., 180 / 255.), (44. / 255., 160 / 255., 44 / 255.), (214 / 255., 39 / 255., 40 / 255.), (255 / 255., 127 / 255., 14 / 255.) ]

ax = plt.subplot(2, 3, 1)
plt.title("a) K03", fontsize = 14)

plt.plot( temp1[:], qc[0], 'o', markerfacecolor = cs[2], markeredgecolor = cs[2], markeredgewidth = 2., markersize = 8)
plt.plot( temp1[:], qi[0], 'D', markerfacecolor = cs[0], markersize = 8, markeredgecolor = cs[0], markeredgewidth = 2. )

plt.legend(["Liquid", "Ice"], frameon = False, ncol = 1, loc = "lower right" )

make_axis( ax )
plt.ylabel( "Cloud Water Path [g / m$^2$]", fontsize = 14)

majorLocator = MultipleLocator(10)
minorLocator = MultipleLocator(5)
ax.yaxis.set_major_locator(majorLocator)
ax.yaxis.set_minor_locator(minorLocator)

plt.ylim([0., 42.5])

ax = plt.subplot(2, 3, 4)
plt.title( "d)", fontsize = 14)

plt.plot( temp1[:], qr[0], 'o', markerfacecolor = 'None', markersize = 8, markeredgecolor = cs[2], markeredgewidth = 2. )
plt.plot( temp1[:], qg[0], 'D', markerfacecolor = 'None', markersize = 8, markeredgecolor = cs[0], markeredgewidth = 2.)
plt.plot( temp1[:], qs[0], '*', markerfacecolor = 'None', markersize = 8, markeredgecolor = cs[0], markeredgewidth = 2., alpha = 0.3 )

plt.legend(["Rain", "Graupel", "Snow"], frameon = False, ncol = 1, loc = "upper left" )

make_axis( ax )
plt.ylabel( "Precipitating Water Path [g / m$^2$]", fontsize = 14)

majorLocator = MultipleLocator(10)
minorLocator = MultipleLocator(5)
ax.yaxis.set_major_locator(majorLocator)
ax.yaxis.set_minor_locator(minorLocator)

plt.ylim([0., 60.])

ax = plt.subplot(2, 3, 2)
plt.title("b) NA5", fontsize = 14)

plt.plot( temp1[:], qc[1], 'o', markerfacecolor = cs[2], markeredgecolor = cs[2], markeredgewidth = 2., markersize = 8 )
plt.plot( temp1[:], qi[1], 'D', markerfacecolor = cs[0], markeredgecolor = cs[0], markeredgewidth = 2., markersize = 8 )

make_axis( ax )
plt.ylabel( "Cloud Water Path [g / m$^2$]", fontsize = 14)

majorLocator = MultipleLocator(10)
minorLocator = MultipleLocator(5)
ax.yaxis.set_major_locator(majorLocator)
ax.yaxis.set_minor_locator(minorLocator)

plt.ylim([0., 42.5])

ax = plt.subplot(2, 3, 5)
plt.title("d)")

plt.plot( temp1[:], qr[1], 'o', markerfacecolor = 'None', markeredgecolor = cs[2], markeredgewidth = 2., markersize = 8)
plt.plot( temp1[:], qg[1], 'D', markerfacecolor = 'None', markersize = 8, markeredgecolor = cs[0], markeredgewidth = 2.)
plt.plot( temp1[:], qs[1], '*', markerfacecolor = 'None', markersize = 8, markeredgecolor = cs[0], markeredgewidth = 2., alpha = 0.3 )

make_axis( ax )
plt.ylabel( "Precipitating Water Path [g / m$^2$]", fontsize = 14)

majorLocator = MultipleLocator(10)
minorLocator = MultipleLocator(5)
ax.yaxis.set_major_locator(majorLocator)
ax.yaxis.set_minor_locator(minorLocator)

plt.ylim([0., 60.])

ax = plt.subplot(2, 3, 3)
plt.title("c) M05", fontsize = 14)

plt.plot( temp1[:], qc[2], 'o', markerfacecolor = cs[2], markeredgecolor = cs[2], markeredgewidth = 2., markersize = 8 )
plt.plot( temp1[:], qi[2], 'D', markerfacecolor = cs[0], markeredgecolor = cs[0], markeredgewidth = 2., markersize = 8 )


make_axis( ax )
plt.ylabel( "Cloud Water Path [g / m$^2$]", fontsize = 14)

majorLocator = MultipleLocator(10)
minorLocator = MultipleLocator(5)
ax.yaxis.set_major_locator(majorLocator)
ax.yaxis.set_minor_locator(minorLocator)

plt.ylim([0., 42.5])

ax = plt.subplot(2, 3, 6)
plt.title( "f)", fontsize = 14)

plt.plot( temp1[:], qr[2], 'o', markerfacecolor = 'None', markeredgecolor = cs[2], markeredgewidth = 2., markersize = 8)
plt.plot( temp1[:], qg[2], 'D', markerfacecolor = 'None', markersize = 8, markeredgecolor = cs[0], markeredgewidth = 2.)
plt.plot( temp1[:], qs[2], '*', markerfacecolor = 'None', markersize = 8, markeredgecolor = cs[0], markeredgewidth = 2., alpha = 0.3 )

make_axis( ax )
plt.ylabel( "Precipitating Water Path [g / m$^2$]", fontsize = 14)

majorLocator = MultipleLocator(10)
minorLocator = MultipleLocator(5)
ax.yaxis.set_major_locator(majorLocator)
ax.yaxis.set_minor_locator(minorLocator)

plt.ylim([0., 60.])

plt.show()
```

```python
#Now plot profiles
cloud = np.load( "data/small_domain_micro_cloud.dat") #cloud fraction
qncloud = np.load( "data/small_domain_micro_qncloud.dat") #cloud condensate mixing ratio, averaged over cloudy regions only
rho = np.load( "data/small_domain_micro_rho.dat") #density
qpsrc = np.load( "data/small_domain_micro_qpsrc.dat") * rho #precipitation source

temp = np.load( "data/small_domain_micro_tabs.dat") #plot versus temperature

#mask heights with no clouds
cloud = np.ma.masked_where( cloud < 0., cloud ) 
qncloud = np.ma.masked_where( qncloud < 0., qncloud )
```

```python
def make_vert_axis( a ):

    a.tick_params( axis = 'x', which = 'both', bottom="on", top = "off", labelbottom = "on", direction = 'out')                  
    a.tick_params( axis = 'y', which = 'both', bottom="off", top="off",  labelbottom="off", left="on", right="off", labelleft="on", direction = 'out')                                                                                          

    a.tick_params(length = 5, which = 'both')

    a.spines["top"].set_visible(False)    
    a.spines["right"].set_visible(False)      

    majorLocator = MultipleLocator( 20 )
    majorFormatter = FormatStrFormatter('%d')
    minorLocator = MultipleLocator( 5 )

    a.yaxis.set_major_locator(majorLocator)
    a.yaxis.set_major_formatter(majorFormatter)
    a.yaxis.set_minor_locator(minorLocator)
        
    plt.ylabel( "Temperature [K]", fontsize = 12 )

    plt.axhline( y = 273., color = 'k' )
    plt.ylim([315., 190.])
        
    return 0
```

```python
fig = plt.figure( figsize = (15, 12) )
plt.subplots_adjust(left = 0.07, right = 0.97, bottom = 0.07, top = 0.97, hspace = 0.3, wspace = 0.3)

ax = plt.subplot(3, 3, 1)
plt.title("a) K03")

ax.set_color_cycle([plt.cm.coolwarm(i) for i in np.linspace(0, 1, t)])

for i in range( t ):
    plt.plot( qncloud[0, i], temp[0, i], linewidth = 2., label = stemp1[i] )

plt.xlabel( "Cloud condensate mixing ratio [g kg$^{-1}$]", fontsize = 12 )
plt.xlim([0, 1.])

majorLocator = MultipleLocator( 0.2 )
minorLocator = MultipleLocator( 0.1 )

ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_minor_locator(minorLocator)

make_vert_axis( ax )

ax = plt.subplot(3, 3, 2)
plt.title( "b)")

ax.set_color_cycle([plt.cm.coolwarm(i) for i in np.linspace(0, 1, t)])

for i in range( t ):
    plt.plot( qpsrc[0, i], temp[0, i], linewidth = 2., label = stemp1[i] + 'K' )

plt.xlabel( "Precipitation Source [g m$^{-3}$ day$^{-1}$]", fontsize = 12 )
plt.xlim([0, 1.6])

majorLocator = MultipleLocator( 0.2 )
minorLocator = MultipleLocator( 0.1 )

ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_minor_locator(minorLocator)

make_vert_axis( ax )

ax = plt.subplot(3, 3, 3)
plt.title( "c)")

ax.set_color_cycle([plt.cm.coolwarm(i) for i in np.linspace(0, 1, t)])

for i in range( t ):
    plt.plot( cloud[0, i], temp[0, i], linewidth = 2., label = stemp1[i] + 'K' )

plt.legend(loc = "lower right", frameon = False )

plt.xlabel( "Cloud cover", fontsize = 12 )
plt.xlim([0, .4])

majorLocator = MultipleLocator( 0.1 )
minorLocator = MultipleLocator( 0.05 )

ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_minor_locator(minorLocator)

make_vert_axis( ax )

ax = plt.subplot(3, 3, 4)
plt.title("d) NA5")
ax.set_color_cycle([plt.cm.coolwarm(i) for i in np.linspace(0, 1, t)])

for i in range( t ):
    plt.plot( qncloud[1, i], temp[1, i], linewidth = 2., label = stemp1[i] )

plt.xlabel( "Cloud condensate mixing ratio [g kg$^{-1}$]", fontsize = 12 )
plt.xlim([0, 1.2])

majorLocator = MultipleLocator( 0.2 )
minorLocator = MultipleLocator( 0.1 )

ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_minor_locator(minorLocator)

make_vert_axis( ax )

ax = plt.subplot(3, 3, 5)
plt.title( "e)")

ax.set_color_cycle([plt.cm.coolwarm(i) for i in np.linspace(0, 1, t)])

for i in range( t ):
    plt.plot( qpsrc[1, i], temp[1, i], linewidth = 2., label = stemp1[i] )

plt.xlabel( "Precipitation Source [g m$^{-3}$ day$^{-1}$]", fontsize = 12 )
plt.xlim([0, 1.6])

majorLocator = MultipleLocator( 0.2 )
minorLocator = MultipleLocator( 0.1 )

ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_minor_locator(minorLocator)

make_vert_axis( ax )

ax = plt.subplot(3, 3, 6)
plt.title( "f)")

ax.set_color_cycle([plt.cm.coolwarm(i) for i in np.linspace(0, 1, t)])

for i in range( t ):
    plt.plot( cloud[1, i], temp[1, i], linewidth = 2., label = stemp1[i] )

plt.xlabel( "Cloud cover", fontsize = 12 )
plt.xlim([0, .4])

majorLocator = MultipleLocator( 0.1 )
minorLocator = MultipleLocator( 0.05 )

ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_minor_locator(minorLocator)

make_vert_axis( ax )

ax = plt.subplot(3, 3, 7)
plt.title("g) M05")
ax.set_color_cycle([plt.cm.coolwarm(i) for i in np.linspace(0, 1, t)])

for i in range( t ):
    plt.plot( qncloud[2, i], temp[2, i], linewidth = 2., label = stemp1[i] )

plt.xlabel( "Cloud condensate mixing ratio [g kg$^{-1}$]", fontsize = 12 )
plt.xlim([0, 1.2])

majorLocator = MultipleLocator( 0.2 )
minorLocator = MultipleLocator( 0.1 )

ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_minor_locator(minorLocator)

make_vert_axis( ax )

ax = plt.subplot(3, 3, 8)
plt.title( "h)")

ax.set_color_cycle([plt.cm.coolwarm(i) for i in np.linspace(0, 1, t)])

for i in range( t ):
    plt.plot( qpsrc[2, i], temp[2, i], linewidth = 2., label = stemp1[i] )

plt.xlabel( "Precipitation Source [g m$^{-3}$ day$^{-1}$]", fontsize = 12 )
plt.xlim([0, 1.6])

majorLocator = MultipleLocator( 0.2 )
minorLocator = MultipleLocator( 0.1 )

ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_minor_locator(minorLocator)

make_vert_axis( ax )

ax = plt.subplot(3, 3, 9)
plt.title( "i)")

ax.set_color_cycle([plt.cm.coolwarm(i) for i in np.linspace(0, 1, t)])

for i in range( t ):
    plt.plot( cloud[2, i], temp[2, i], linewidth = 2., label = stemp1[i] )

plt.xlabel( "Cloud cover", fontsize = 12 )
plt.xlim([0, .4])

majorLocator = MultipleLocator( 0.1 )
minorLocator = MultipleLocator( 0.05 )

ax.xaxis.set_major_locator(majorLocator)
ax.xaxis.set_minor_locator(minorLocator)

make_vert_axis( ax )

plt.show()
```

```python
#Now take vertically-averaged cloud condensate mixing ratios
qn = np.mean( np.ma.masked_where( qncloud < 0, qncloud ) * z, axis = 2 ) / np.mean( z, axis = 2 )

#variable resolution:
qncld2 = np.load( "data/small_domain_res_qncloud.dat")
qn2 = np.mean( np.ma.masked_where( qncld2 < 0, qncld2 ) * z, axis = 2 ) / np.mean( z, axis = 2 )
```

```python
fig = plt.figure( figsize = (10, 6) )
plt.subplots_adjust(left = 0.1, right = 0.97, bottom = 0.12, top = 0.92, hspace = 0.3, wspace = 0.3)


ax = plt.subplot(1, 2, 1)
plt.title( "a)", fontsize = 12 )

plt.plot( temp1, qn[0], 'o', markerfacecolor = 'None', markeredgecolor = 'k', markeredgewidth = 2., markersize = 8)
plt.plot( temp1, qn[1], 'v', markerfacecolor = 'None', markersize = 8, markeredgecolor = 'k', markeredgewidth = 2. )
plt.plot( temp1, qn[2], 'D', markerfacecolor = 'None', markersize = 8, markeredgecolor = 'r', markeredgewidth = 2. )

plt.legend(["K03", "NA5", "M05"], frameon = False, ncol = 1, loc = "upper left" )

make_axis( ax )

majorLocator = MultipleLocator(0.05)
minorLocator = MultipleLocator(0.025)
ax.yaxis.set_major_locator(majorLocator)
ax.yaxis.set_minor_locator(minorLocator)

plt.ylim([0., 0.23])
plt.ylabel( "Cloud Density [g kg$^{-1}$]", fontsize = 12 )

ax = plt.subplot(1, 2, 2)
plt.title( "b)", fontsize = 12 )

plt.plot( temp1, qn2[0], 'o', markerfacecolor = 'None', markeredgecolor = 'k', markeredgewidth = 2., markersize = 8)
plt.plot( temp1, qn2[1], 'v', markerfacecolor = 'None', markersize = 8, markeredgecolor = 'k', markeredgewidth = 2. )
plt.plot( temp1, qn2[2], 's', markerfacecolor = 'None', markersize = 8, markeredgecolor = 'k', markeredgewidth = 2. )

plt.legend(["$\Delta x$ = 0.5km", "$\Delta x$ = 1km", "$\Delta x$ = 2km"], frameon = False, ncol = 1, loc = "upper left" )

make_axis( ax )

majorLocator = MultipleLocator(0.05)
minorLocator = MultipleLocator(0.025)
ax.yaxis.set_major_locator(majorLocator)
ax.yaxis.set_minor_locator(minorLocator)

plt.ylim([0., 0.23])
plt.ylabel( "Cloud Density [g kg$^{-1}$]", fontsize = 12 )

plt.show()
```
